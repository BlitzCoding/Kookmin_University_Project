```python
# Task 3 - 1

df = df[['location','date','total_deaths_per_million','total_cases_per_million',
         'people_vaccinated_per_hundred','human_development_index']]

df
```

```python
# Task 3 - 2

recent_group = df.groupby(df['location']).max()

recent_group
```

```python
# Task 3 - 3

recent_group['total_relative_death'] = recent_group['total_deaths_per_million'] / recent_group['total_cases_per_million']

plt.scatter(recent_group['people_vaccinated_per_hundred'], recent_group['total_relative_death'])
```

```python
# Task 3 - 4

x = recent_group['people_vaccinated_per_hundred'].interpolate(methoid = 'linear', limit_direction = 'forward')
y = recent_group['total_relative_death'].interpolate(methoid = 'linear', limit_direction = 'forward')

def Pearson(X, Y):
    mean_X = X.mean()
    mean_Y = Y.mean()
    std_X = (sum((x - mean_X) ** 2 for x in X)/len(X)) ** 0.5
    std_Y = (sum((y - mean_Y) ** 2 for y in Y)/len(Y)) ** 0.5
    cov = sum((x - mean_X) * (y - mean_Y) for x, y in zip(X,
    Y))/len(X)
    corr = cov/std_X/std_Y
    
    return corr

print("Pearson Correlation : ", Pearson(x, y))
```

```python
# Task 4 - 1

x = recent_group['human_development_index'].interpolate(methoid = 'linear', limit_direction = 'forward')
y = recent_group['people_vaccinated_per_hundred'].interpolate(methoid = 'linear', limit_direction = 'forward')

Pearson(x, y)
```

```python
# Task 4 - 2

import torch
import torch.optim as optim

x_train = torch.Tensor([x])
x_train = x_train.squeeze(1)
x_train = x_train.permute(1,0)
y_train = torch.Tensor([y])
y_train = y_train.squeeze(1)
y_train = y_train.permute(1,0)

W = torch.zeros(1, requires_grad = True)

b = torch.zeros(1, requires_grad = True)

for epoch in range(3001):
    hypothesis = x_train * W + b
    cost = torch.mean((hypothesis - y_train) ** 2)
    optimizer = optim.SGD([W, b], lr=0.01)

    optimizer.zero_grad()
    cost.backward() 
    optimizer.step() 
    
    if epoch % 100 == 0:
        print( 'epoch: {}, cost: {:.6f}, W: {}, b: {}'.format(epoch,
            cost.item(), W.squeeze(), b))
```
